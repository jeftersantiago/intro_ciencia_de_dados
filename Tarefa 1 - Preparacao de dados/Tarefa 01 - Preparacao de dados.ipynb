# Exercicio 2
Leia os dados com erro novamente e insira a mediana de cada atributo onde for encontrada `NaN`.

```python
data = pd.read_csv('data/iris-with-errors.csv', header=(0))
data = data.replace('?', np.nan)

X = np.array(data[data.columns[0:data.shape[1] - 1]], dtype=float)
medians = np.nanmedian(X, axis = 0)

for i in np.arange(0, X.shape[0]):
  for j in np.arange(0, X.shape[1]):
    if(np.isnan(X[i, j])):
      X[i, j] = medians[j]

print(X)
```

# Exercício 3
Conside os dados da Iris e mostre a distribuição de probabilidades de cada uma das variáveis após a normalização e padronização.

```python
data = pd.read_csv('data/iris.csv', header=(0))

X = np.array(data[data.columns[0:data.shape[1] - 1]])
```

###  Normaliza os dados no intervalo de [0, 1]

```python

scaler = MinMaxScaler(feature_range = (0, 1))

X_norm = scaler.fit_transform(X)

print(X_norm)
```

### Padronizando os dados 
 - Média = 0 
 - Variância = 1

```python
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)

print(rescaledX)
```

```python
print()
for i in np.arange(0,rescaledX.shape[1]):
    print('Media da coluna', i, '(', data.columns[i], ') é igual a: ', np.mean(rescaledX[:,i]))
    print('Desvio padrão da coluna', i,'(',  data.columns[i], ') é igual a: ', np.std(rescaledX[:,i]),'\n')
```

## Encontrando outliers para a base Iris

```python
data = pd.read_csv('data/iris.csv', header=(0))
data = data.drop(data.columns[[-2, -1]], axis = 1)

data[data > 150] = data[data > 150]*10

print(data)
```

```python
sns.pairplot(data)
plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

print (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))))
```

```python
print (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1))
```

```python
v = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)
data = data.drop(data.index[list(v)], axis=0)

print(data.head(5))
print(data.shape)
```

Foi removido do DataFrame os outliers, note que além do shape do DataFrame estar diferente também podemos refazer o plot do gráfico e verá que não tem mais a presença dos outliers.

```python
sns.pairplot(data)
plt.show()
```
